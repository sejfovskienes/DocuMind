```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
```

```python
text_chunks = [
    """
    Tim Cook in visit of Cupertino Headquarters
    Apple Inc. announced yesterday that CEO Tim Cook will visit their Cupertio headquarters next Monday.
    The tech giant plans to unveil the new iPhone 15 at an event in San Francisco. 
    """,
    """
    Microso􀅌 and Google are also competing in the artificial intelligence market. 
    According to Bloomberg, the company's stock rose 3.5% on NASDAQ.
    Dr. Sarah Johnson from Stanford University commented on the innovation. 
     """,
     """ 
    The European Union has imposed regula􀆟ons, while President Biden met with Prime Minister 
    Rishi Sunak in Washington D.C. Amazon reported $500 million in quarterly revenue. 
    Tesla's Elon Musk tweeted about the developments from Aus􀆟n, Texas.
     """
]
```

```python
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
for i, chunk in enumerate(text_chunks):
    index=i+1
    print(f"{index}TH CHUNK:\n{chunk}")
    ner_results = nlp(str(chunk))
    print(f"\t NER RESULTS FROM {index}TH CHUNK:\n")
    for result in ner_results:
        print(result)
```
