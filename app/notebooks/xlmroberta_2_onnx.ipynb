```python
import torch
import pathlib
from transformers.onnx import export, FeaturesManager
from transformers import AutoTokenizer, AutoModelForTokenClassification
```

```python
MODEL_NAME = "FacebookAI/xlm-roberta-base"
ONNX_EXPORT_PATH = pathlib.Path("../core/onnx/xlm_roberta_ner.onnx")
ONNX_EXPORT_PATH.parent.mkdir(parents=True, exist_ok=True) 
```

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForTokenClassification.from_pretrained(MODEL_NAME)
```

```python
onnx_config_class = FeaturesManager.get_config(model_type=model.config.model_type, feature="token-classification")
onnx_config = onnx_config_class(model.config)
```

```python
export(
    preprocessor=tokenizer,
    model=model,
    config=onnx_config,
    opset=13,
    output=ONNX_EXPORT_PATH
)

print(f"{MODEL_NAME} successfully exported to: {ONNX_EXPORT_PATH}")
```
